The triclusters have been filtered by:  
- Bonferroni  
- Significative number of links that induced a split coherent to the cluster they have been assigned to (with respect to a Fisher test)  
- 13-clusters have been removed  
  
7697 rows are left. Each of these has been enriched via a hypergeometric gsea.  


```r
nrow(filtered_triclusters)
filtered_triclusters[1,]
```

All the 7697 triclusters have been enriched using kegg, using reactome and using go.  
The p-values related to each database have been adjusted via BH procedure (independently for kegg, reactome and go).  
  
- pval: the pval of the pathway in the hypergeometric test, not adjusted  
- pval_BH_adj: the pval after BH adjustment (each database adjusted independently)  
- size: number of genes given to the test  
- overlap: number of genes present in the pathway
- overlapGenes: Entrez IDs of the genes present in the pathway

For inspecting the results I provide the functions _get_graph_of_row_ and and _get_graph_of_multiple_rows_.  
- ARG 1: idx of the row / list of idxs of multiple rows
- ARG 2: include_zeros: if TRUE not only the 1-valued edges are shown (in black), but also the 0-valued edges (in red)

```r
g <- get_graph_of_row(89, include_zeros=TRUE)
plot(g)
```
